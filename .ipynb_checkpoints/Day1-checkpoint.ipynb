## Part 1

```python
with open('day1_input.txt','r') as f:
    line = f.read()

line += line[0]

print(line)
matches = [int(c) for i, c in enumerate(line) if i<(len(line)-1) and c==line[i+1]]
print(matches)
print(sum(matches))
```

## Part 2

```python
with open('day1_input.txt','r') as f:
    line = f.read()

step = len(line) / 2
matches = [int(c) for i, c in enumerate(line) if c==line[int((i+step)%len(line))]]
print(matches)
print(sum(matches))
```

```python

    1212 produces 6: the list contains 4 items, and all four digits match the digit 2 items ahead.
    1221 produces 0, because every comparison is between a 1 and a 2.
    123425 produces 4, because both 2s match each other, but no other digit has a match.
    123123 produces 12.
    12131415 produces 4.
```

```python
lines = ['1212','1221','123425','123123','12131415']
for line in lines:
    step = len(line) / 2
#     print(len(line))
#     print(step)
    matches = [int(c) for i, c in enumerate(line) if c==line[int((i+step)%len(line))]]
    print(matches)
    print(sum(matches))
```
